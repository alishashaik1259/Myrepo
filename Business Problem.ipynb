# Introduction / Business Problem

 A multinational chain of coffeehouses is looking for a right location in Toronto, Canada, where they can kickstart their services like Café’s, Coffee Houses, etc and further expand. As a Data Scientist, depending on the various venues in the neighbourhoods of Toronto, we have to come up with the location for the Café/Coffee house where the business will be successful and there is opportunity for expansion further.

# Data 

For this particular problem, the data we need is geographical locations of various popular venues in the neighbourhoods of Toronto, Canada. The venues are of different types, for example, Café’s, Restaurants, Bars, Pizza places, Pharmacies, Pubs, Bagel Shops, etc. Geographical location values are nothing but, the latitude and longitude values. We can get these values from Foursquare API. Depending on this data, the scope for a launch of a Café’s/Coffee House in a right location is determined.
